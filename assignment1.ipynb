### Problem 1

##### Initial State
* All 64 discs are stacked in order from largest to smallest diameter on a single pole.

##### Goal state
* All 64 discs are stacked on a different pole than they started on in the same order they started in.

##### Restrictions
* Only one disc can be moved at a time.
* A disc may only be place on top of a larger disc or on an empty pole.

##### States
* Each state is a valid configureation of the 64 discs across the 3 poles.
* A state can be defined by the ordered list of discs on each pole
* [1,...,64] [] [] -> discs 1 to 64 are on pole 1, no discs are on poles 2 or 3

##### Actions
* The only action is moving the top disc from one pole, onto another pole, assuming there is a larger disc at the top of the new pole.

##### State Space Diagram
* Nodes which contain already seen states will not be explicitly expanded
* Diagram drawn using draw.io



##### Sequence of Actions
* This is an example of solving a 4 disk tower of hanio puzzle, however the same mechanics apply to larger puzzles.
* A larger puzzle has many more steps and deeper levels of recursion.
* Source: https://mathworld.wolfram.com/TowerofHanoi.html

```python
def generateMoves(maze, pos):
    moves = []
    x = pos[0]
    y = pos[1]

    if x > 0 and maze[y][x - 1] == 0:
        moves.append((x - 1, y))

    if x < 24 and maze[y][x + 1] == 0:
        moves.append((x + 1, y))

    if y > 0 and maze[y - 1][x] == 0:
        moves.append((x, y - 1))

    if y < 24 and maze[y + 1][x] == 0:
        moves.append((x, y + 1))

    return moves
```

```python
from collections import deque

PATH = 0.5
EXPLORED = 0.2

def BFS(start, goal):
    maze = getCleanMaze()

    # Create a queue for BFS
    queue = deque([start])

    # Track parent for generating path later
    parent = {}
    parent[start] = None
    nodesExpanded = 0

    while queue:
        node = queue.popleft()
        nodesExpanded += 1

        if node == goal:
            break

        moves = generateMoves(maze, node)
        for move in moves:
            # Only visit unvisited nodes
            if move not in parent:
                queue.append(move)
                parent[move] = node

        maze[node[1]][node[0]] = EXPLORED

    # Reconstruct path from start to goal
    path = []
    current = goal
    while current:
        path.append(current)
        maze[current[1]][current[0]] = PATH
        current = parent[current]
    path.reverse()

    print(f"Cost: {len(path)}")
    print(f"Nodes Expanded: {nodesExpanded}")
    
    plotMaze(maze)
```

#### BFS - Deliverable 1
* At each location, based on the movement constraints and the valid moves left in the maze, a list of the next possible moves is generated
* The moves are added to the exploration queue in the order they are discovered to maintain a breadth first search
* Using a parent dictionary to track the revealed path.

```python
def DFS(start, goal):
    maze = getCleanMaze()

    # Create a queue for BFS
    queue = deque([start])

    # Track parent for generating path later
    parent = {}
    parent[start] = None
    nodesExpanded = 0

    while queue:
        node = queue.popleft()
        nodesExpanded += 1

        if node == goal:
            break

        moves = generateMoves(maze, node)
        for move in moves:
            # Only visit unvisited nodes
            if move not in parent:
                queue.appendleft(move)
                parent[move] = node

        maze[node[1]][node[0]] = EXPLORED

    # Reconstruct path from start to goal
    path = []
    current = goal
    while current:
        path.append(current)
        maze[current[1]][current[0]] = PATH
        current = parent[current]
    path.reverse()

    print(f"Cost: {len(path)}")
    print(f"Nodes Expanded: {nodesExpanded}")
    
    plotMaze(maze)
```

#### DFS - Deliverable 1
* Same implementation except instead of appending to the queue, new moves are pre-pended to the queue so the algorithm searches deep first

```python
import heapq

def manhattanHeuristic(node, goal):
    return abs(node[0] - goal[0]) + abs(node[1] - goal[1])

def AStar(start, goal):
    maze = getCleanMaze()

    # Priority queue for A*
    queue = []
    heapq.heappush(queue, (0, start))

    # Track costs and path
    gCosts = {start: 0}
    parent = {start: None}
    nodesExpanded = 0

    while queue:
        _, node = heapq.heappop(queue)
        nodesExpanded += 1

        if node == goal:
            break

        # Generate possible moves
        moves = generateMoves(maze, node)
        for move in moves:
            newGCost = gCosts[node] + 1

            if move not in gCosts or newGCost < gCosts[move]:
                # Update g(n) for the move
                gCosts[move] = newGCost
                # f(n) = g(n) + h(n)
                fCost = newGCost + manhattanHeuristic(move, goal)
                heapq.heappush(queue, (fCost, move))
                parent[move] = node

        maze[node[1]][node[0]] = EXPLORED

    # Reconstruct path from start to goal
    path = []
    current = goal
    while current:
        path.append(current)
        maze[current[1]][current[0]] = PATH
        current = parent[current]
    path.reverse()

    print(f"Cost: {len(path)}")
    print(f"Nodes Expanded: {nodesExpanded}")
    
    plotMaze(maze)
```

#### A* Deliverable 1
* Used the same framework as BFS and DFS to implement A*.
    * Same move generator funciton.
* Main difference is the queue is now priority managed by heapq.
* The priority of the queue is determined by the f(n) of a move.
* When a new set of potential moves is generated, its path cost (g(n)) is calculated.
    * If there path cost for this position has been calculated before, update it if the new one is less.
* Then the cost fucntion is evaluated, which in this case is the path cost plus the manhattan distance (heuristic) (f(n) = g(n) + h(n)).
* We add items to the queue along with their cost and heapq will add the item to the queue in order of ascending cost function.
* We expand nodes by popping positions off the queue.

##### Why Manhattan?
* The only information our algorithm has about the maze is the position of the start and the position of the goal.
* Without knowing anything about the obstacles or the shape of the maze, using a distance based heuristic is logical.
* Could have chosed chosen to use the Eclidean distance and that would also have been valid.
* Both hueristics are admissble and suitable for use with A*.
* I chose Manhattan becasue I found it to be more representative given the constrains on our movement.
* Eclidean distacne computes the minimum diagonal distance between two points whereas Manhattan is the sum of the x and y distances.
* Since we cannot move diagonally, I chose to use Manhattan distance as my heuristic.
